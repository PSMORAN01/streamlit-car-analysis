```python
import pandas as pd
import plotly.express as px

# Simulación del DataFrame df
data = {
    "Brand": ["Kia", "Chevrolet", "Mercedes", "Audi", "Volkswagen"],
    "Model": ["Rio", "Malibu", "GLA", "Q5", "Golf"],
    "Year": [2020, 2012, 2020, 2023, 2003],
    "Engine_Size": [4.2, 2.0, 4.2, 2.0, 2.6],
    "Fuel_Type": ["Diesel", "Hybrid", "Diesel", "Electric", "Hybrid"],
    "Transmission": ["Manual", "Automatic", "Automatic", "Manual", "Semi-Automatic"],
    "Mileage": [289944, 5356, 231440, 160971, 286618],
    "Doors": [3, 2, 4, 2, 3],
    "Owner_Count": [5, 3, 2, 1, 3],
    "Price": [8501, 12092, 11171, 11780, 2867]
}
df = pd.DataFrame(data)

# Gráfico 1: Cantidad de Vehículos por Tipo de Combustible
fig1 = px.histogram(df, x="Fuel_Type", title="Cantidad de Vehículos por Tipo de Combustible", labels={"Fuel_Type": "Tipo de Combustible"}, color="Fuel_Type")
fig1.show()

# Gráfico 2: Distribución de Precios de Vehículos
fig2 = px.histogram(df, x="Price", title="Distribución de Precios de Vehículos", labels={"Price": "Precio"}, nbins=10, color_discrete_sequence=["skyblue"])
fig2.show()

# Gráfico 3: Relación entre Kilometraje y Año del Vehículo
fig3 = px.scatter(df, x="Year", y="Mileage", color="Fuel_Type", title="Relación entre Kilometraje y Año del Vehículo", labels={"Year": "Año del Vehículo", "Mileage": "Kilometraje"})
fig3.show()
```
